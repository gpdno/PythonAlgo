```python
de = pd.read_csv("DE_data")
```

```python
de.tail()
```

```python
de['Date'] = pd.to_datetime(de['Date'])
```

```python
de.set_index('Date', inplace=True)
```

```python
def take_first(array_like):
    return array_like[0]

def take_last(array_like):
    return array_like[-1]

de_weekly = de.resample('W',                                 # Weekly resample
                    how={'Adj. Open': take_first, 
                         'Adj. High': 'max',
                         'Adj. Low': 'min',
                         'Adj. Close': take_last,
                         'Adj. Volume': 'sum'}, 
                    loffset=pd.offsets.timedelta(days=-6)) # to put the labels to Monday

de_weekly = de_weekly[['Adj. Open', 'Adj. High', 'Adj. Low', 'Adj. Close', 'Adj. Volume']]
```
